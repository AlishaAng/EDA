```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('../Medical_prognosis/Data/heart.csv')
```

### Histogram

```python
sns.set_palette("bright")
sns.histplot(data=df, x="age", hue='sex')
plt.show()
```

```python
sns.set_palette("bright")
sns.histplot(data=df, x="age", hue='sex', kde=True)
plt.show()
```

### Rug plot

```python
# Subset to females
female_subset = df[df['sex'] == 0]

# Set plot style
sns.set_style('whitegrid')

# Create a density plot with a rug plot
sns.kdeplot(data=female_subset['age'], color='darkblue', linewidth=3, label='Age Distribution')
sns.rugplot(data=female_subset['age'], color='black', height=0.05)

# Set plot title and axis labels
plt.title('Density Plot with Rug Plot for Female patients')
plt.xlabel('Age')
plt.ylabel('Density')
plt.legend()
plt.show()
```

### Density plots

```python
sns.kdeplot(data=df, x= 'age', hue='sex', fill=True, common_norm=False, palette='Set1')
plt.show()
```
