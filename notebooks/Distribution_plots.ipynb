### Visualising Distributions

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("../data/Sleep_health_and_lifestyle_dataset.csv").iloc[:,1:]
```

```python
df.head()
```

```python
# Fill null values in the 'Sleep Disorder' column with the string "No disorder"
# Assuming that empty value means patient does not have a sleep disorder 
df['Sleep Disorder'] = df['Sleep Disorder'].fillna('No disorder')
```

### Histogram

```python
sns.set_palette("bright")
sns.histplot(data=df, x="Sleep Duration", hue='Sleep Disorder', kde=True)
plt.show()
```

### Density plots

```python
sns.kdeplot(data=df, x= 'Sleep Duration', hue='Sleep Disorder', fill=True, common_norm=False, palette='Set1')
plt.show()
```

### Rug plot

```python
# Subset
subset = df[df['Sleep Disorder'] == 'Insomnia']

# Set plot style
sns.set_style('whitegrid')

# Create a density plot with a rug plot
sns.kdeplot(data=subset['Sleep Duration'], color='darkblue', linewidth=3, label='Sleep Duration Distribution')
sns.rugplot(data=subset['Sleep Duration'], color='black', height=0.05)

# Set plot title and axis labels
plt.title('Density Plot with Rug Plot for Insomnia')
plt.xlabel('Sleep Duration')
plt.ylabel('Density')
plt.legend()
plt.show()
```
