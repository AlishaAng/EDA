### Visualising Distributions

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = sns.load_dataset("healthexp")
```

```python
df.head()
```

### Histogram

```python
sns.set_palette("bright")
sns.histplot(data=df, x="Life_Expectancy", hue='Country', kde=True)
plt.show()
```

### Density plots

```python
sns.kdeplot(data=df, x= 'Life_Expectancy', hue='Country', fill=True, common_norm=False, palette='Set1')
plt.show()
```

### Rug plot

```python
# Subset
subset = df[df['Country'] == 'Japan']

# Set plot style
sns.set_style('whitegrid')

# Create a density plot with a rug plot
sns.kdeplot(data=subset['Life_Expectancy'], color='darkblue', linewidth=3, label='Life Expectancy Distribution')
sns.rugplot(data=subset['Life_Expectancy'], color='black', height=0.05)

# Set plot title and axis labels
plt.title('Density Plot with Rug Plot for Japan')
plt.xlabel('Life Expectancy')
plt.ylabel('Density')
plt.legend()
plt.show()
```
